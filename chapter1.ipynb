```python
import gym
from gym import envs
import time
from gym.spaces import Discrete, Box
from IPython import display 
import PIL.Image  

# # 创建环境时指定 render_mode='rgb_array'
env = gym.make('CartPole-v1',render_mode='rgb_array')

env.reset()
img = PIL.Image.fromarray(env.render()) 
display.display(img)

while True:
    action  = env.action_space.sample()

    observation, reward, done,truncated, info =env.step(action)
    # 渲染当前帧  
    img = PIL.Image.fromarray(env.render()) 
    display.clear_output(wait=True)  # 清除之前的输出  
    display.display(img)  
    time.sleep(0.05)  # 添加小延时使显示更流畅
    if done:
        break

# 在env.reset()或env.step()后，可以用以下语句以图形化的方法显示当前环境。
# env.render()

env.close()
# 注意：如果绘制了实验的图形界面窗口，那么关闭该窗口的最佳方式是调用env.close()。
# 试图直接关闭图形界面窗口可能会导致内存不能释放，甚至会导致死机

```
